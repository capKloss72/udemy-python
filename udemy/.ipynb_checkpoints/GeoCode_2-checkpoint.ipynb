```python
import pandas
import xlrd
import geopy
from geopy import geocoders
from geopy.geocoders import Nominatim
df_csv = pandas.read_csv("supermarkets.csv")
df_json = pandas.read_json("supermarkets.json")
df_excel = pandas.read_excel("supermarkets.xlsx", sheet_name=0)
df_comma_txt = pandas.read_csv("supermarkets-commas.txt")
df_semi_txt = pandas.read_csv("supermarkets-semi-colons.txt", sep=";")
df_web_csv = pandas.read_csv("http://pythonhow.com/supermarkets.csv")
df_web_json = pandas.read_json("http://pythonhow.com/supermarkets.json")
df_json = df_json.set_index("Address")
nom = Nominatim()
```

```python
df_json
```

```python
df_long_addr = df_csv
```

```python
df_long_addr["Address"]=df_web_json["Address"]+", "+df_web_json["City"]+", "+df_web_json["State"]+", "+df_web_json["Country"]
df_long_addr.set_index("Address")
df_long_addr
```

```python
df_long_addr.set_index("Address")
df_long_addr["Coordinates"]=df_long_addr["Address"].apply(nom.geocode)
df_long_addr
```

```python
df_long_addr["Latitude"]=df_long_addr["Coordinates"].apply(lambda x : x.latitude if x != None else None)
df_long_addr["Longitude"]=df_long_addr["Coordinates"].apply(lambda x : x.longitude  if x != None else None)
df_long_addr
```
